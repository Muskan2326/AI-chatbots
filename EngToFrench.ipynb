```python
!pip install --quiet transformers sentencepiece accelerate

from transformers import pipeline
from IPython.display import display, HTML, clear_output
import time

translator = pipeline('translation_en_to_fr', model='Helsinki-NLP/opus-mt-en-fr')
print("✅ Translator loaded successfully! Ready to translate.")

def typing_effect(text, delay=0.03):
    """Simulate a typing effect in the notebook output."""
    clear_output(wait=True)
    display(HTML("<div style='font-family:monospace; font-size:16px; white-space:pre-wrap;'></div>"))
    out = ''
    for ch in text:
        out += ch
        clear_output(wait=True)
        display(HTML(f"<div style='font-family:monospace; font-size:16px; white-space:pre-wrap;'>{out}</div>"))
        time.sleep(delay)
    print()

def translate_en_to_fr(text):
    """Translate English text to French using Hugging Face pipeline."""
    result = translator(text, max_length=512)
    return result[0]['translation_text']

print("🤖 English → French Chatbot\n(Type 'exit' to quit)\n")

while True:
    src = input("🧑 You (English): ").strip()
    if src.lower() in ("exit", "quit"):
        print("👋 Goodbye!")
        break
    if not src:
        continue
    translation = translate_en_to_fr(src)
    typing_effect(f"🇫🇷 Chatbot: {translation}")
```
